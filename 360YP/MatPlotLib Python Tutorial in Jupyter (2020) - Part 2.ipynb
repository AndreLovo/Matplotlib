```python
#9º Vamos fazer o mesmo gráfico mas agora pegando os valores das receitas (Revenue) e clientes (Visitors)

plt.figure(figsize=(15,10)) #Aumentando  o tamanho do gráfico
plt.plot(a['Visitors'], label="Nº de Clientes")
plt.plot(a['Revenue'], label="Receitas", linestyle="-.")
plt.xticks((range(0,len(a['Visitors']))),a['Date'], rotation = 65)
plt.title("Clientes e Receitas por Dia", fontsize=20)
plt.legend(fontsize=15)
plt.xlabel("Datas")
plt.ylabel("")
plt.show()
```

## 6- Gráfico de Linhas + Gráfico de Barras usando Pandas DFs (Data Frames)

```python
# Vamos usar o gráfico acima
plt.figure(figsize=(15,10)) #Aumentando  o tamanho do gráfico
plt.bar(range (0,len(a['Visitors'])), a['Visitors'], label="Nº de Clientes", color="green")
plt.plot(a['Revenue'], label="Receitas", linestyle="-.")
plt.xticks((range(0,len(a['Visitors']))),a['Date'], rotation = 65)
plt.title("Clientes e Receitas por Dia", fontsize=20)
plt.legend(fontsize=15)
plt.xlabel("Datas")
plt.ylabel("")
plt.show()
```

```python
# Vamos colocar duas barras (montar dois gráficos de barra no mesmo eixo)
# Vamos usar o gráfico acima
plt.figure(figsize=(15,10)) #Aumentando  o tamanho do gráfico
plt.bar(range (0,len(a['Visitors'])), a['Visitors'], label="Nº de Clientes", color="green")
plt.bar(range (0,len(a['Visitors'])), a['Marketing Spend'], label="Gastos com Marketing", color="red")
plt.plot(a['Revenue'], label="Receitas", linestyle="-.")
plt.xticks((range(0,len(a['Visitors']))),a['Date'], rotation = 65)
plt.title("Clientes e Receitas por Dia", fontsize=20)
plt.legend(fontsize=15)
plt.xlabel("Datas")
plt.ylabel("")
plt.show()
```

```python
# Vamos usar o gráfico acima
# Vamos agora usar a coluna Receitas e fazer o gráfico de barras
plt.figure(figsize=(15,10)) #Aumentando  o tamanho do gráfico
plt.bar(range (0,len(a['Revenue'])), a['Revenue'], label="Receitas", color="green")
plt.bar(range (0,len(a['Visitors'])), a['Marketing Spend'], label="Gastos com Marketing", color="red")
plt.plot(a['Visitors'], label="Clientes", linestyle="-.")
plt.xticks((range(0,len(a['Visitors']))),a['Date'], rotation = 65)
plt.title("Clientes e Receitas por Dia", fontsize=20)
plt.legend(fontsize=15)
plt.xlabel("Datas")
plt.ylabel("")
plt.show()
```

```python
# Vamos criar um outro gráfico com 2 barras e 2 escalas

a[["Revenue","Marketing Spend"]].plot(kind="bar", color=["green","red"], figsize=(15,10))
plt.xticks((range(0,len(a['Visitors']))),a['Date'], rotation = 65)
plt.title("Clientes e Receitas por Dia", fontsize=20)
plt.legend(fontsize=15)
plt.xlabel("Datas")
plt.ylabel("")
plt.show()
```

## 6- Gráfico de Dispersão usando Pandas DFs (Data Frames)

```python
# Simples Dispersão - Vamos investigar relacionamentos (Se as variáveis tem influência uma na outra)

plt.figure(figsize=(15,10)) #Tamanho do gráfico
plt.scatter(raw_data["Revenue"], raw_data["Marketing Spend"], color="red", marker="+")
plt.title("Gráfico de Dispersão: Receitas x Gastos com Marketing", fontsize=20)
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Receitas $")
plt.ylabel("Despesas com Marketing")
plt.show()
```

```python
raw_data.head()
```

```python
# Vamos verificar a influencia das promoções nas receitas e nos Gastos com Marketing

groups = raw_data[["Promo","Revenue","Marketing Spend"]].groupby("Promo")
groups
#DataFrameGroupBy

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group['Revenue'], group['Marketing Spend'], marker = 'x', label=promo)
    
ax.legend(loc="upper left")
plt.title("Gráfico de Dispersão: Receitas x Gastos com Marketing através das Promoções", fontsize=20)
plt.xlabel("Receitas")
plt.ylabel("Despesas com Marketing")
    
plt.show()
```

```python
#Vamos fazer a mesma coisa do gráfico acima, mas agora no lugar da promoção vamos colocar as datas

# Vamos verificar a influencia das promoções nas receitas e nos Gastos com Marketing

groups = raw_data[["Day_Name","Revenue","Marketing Spend"]].groupby("Day_Name")
groups
#DataFrameGroupBy

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group['Revenue'], group['Marketing Spend'], marker = 'x', label=promo)
    
ax.legend(loc="upper left")
plt.title("Gráfico de Dispersão: Receitas x Gastos com Marketing através dos Dias", fontsize=20)
plt.xlabel("Receitas")
plt.ylabel("Despesas com Marketing")
    
plt.show()
```
