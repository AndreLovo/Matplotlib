```python
# Importando e instalandos as bibliotecas
import os #fornece funções de interação com o sistema operacional
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## 1- Gráficos de Linhas usando o Matplotlib e Numpy

```python
#1º Para criarmos um gráfico de linhas precisamos criar um array
a = np.array([1,2,3,4,5])
a

#2º Plotando o gráfico
plt.plot(a)#plotar
plt.show()#mostrar
```

```python
# Craindo um segundo Array
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])

# Plotando o gráfico
plt.plot(a,b)#plotar
plt.show()#mostrar
# NÃO É ISSO QUE QUEREMOS
```

```python
# Aumentando o tamanho dos gráficos
plt.figure(figsize=(15,10))

# Colocando os 2 arrays no mesmo gráfico
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
c = np.array([3,1,4,6,4])

# Plotando o gráfico
plt.plot(a, c='green', linestyle='-.', label="AluguelMedio-Ano20")#plotar o array a, alterar a cor -> c ou color, alterar o estilo das linhas,adicionar legenda
# no nome do array
plt.plot(b, linewidth=6., label="Media_dos_Alugueis")#plotar e aumentar a largura ou espessura da linha
plt.plot(c, c='red', linestyle='--', label="AluguelMedio-Ano19")#plotar, alterar a cor -> c ou color, alterar o estilo das linhas
# Adicionando um titulo
plt.title("Resumo dos Alugues por Mês", fontsize=24, fontweight='bold',color='orange')
plt.legend(loc="upper left", fontsize=18)#Para criarmos e mostrarmos as legendas do lado esquerdo superior (localização)
#alterando os eixos
plt.xticks(range (0,len(a)),('Jan','Fev','Mar','Abr','Mai'))# inicio 0 até o tamanho do array a
# Podemos trocar por numeros
# plt.xticks([0,1,2,3,4])
#Colocando os titulos dos eixos
plt.xlabel('Meses')
plt.ylabel('Alugueis Médios')
plt.show()#mostrar
```

## 2- Gráfico de Dispersão

```python
#1-Montando a base de dados
#1º Devemos ter acesso ao banco de dados-.criaremos 2 arrays a e b
#2º Precisamos de eixos x e y

# Vamos criar um array de 30 números
a= np.arange(30)

# Vamos criar um array de 30 numeros randomicos de 0 ate 30
b=np.random.randint(0,30,30) # range de 0 ate 30 e a quantidade é de 30 valores

```

```python
#2- Vamos plotar o gráfico
plt.scatter(a,b)#plotar
plt.show()#mostrar
```

```python
#3- Melhorando o grafico de dispersão
# Vamos criar um outro array de 30 numeros randomicos de 0 ate 30
s=np.random.randint(0,100,30) # range de 0 ate 30 e a quantidade é de 30 valores

#Vamos manipular os gráficos para deixarmos mais relações entre os arrays
b= a + np.random.randint(0,30,30)

plt.scatter(a,b, color='green', s=s)#plotar, colocar cor, alterar tamanho do ponto dos valores randomicos
plt.show()#mostrar
```

## 3- Gráfico de Barras Usando o Matplotlib e Numpy

```python
#1º Para trabalhamros com o Gráfico de Barras precisamos acertar o nosso banco de dados
#2º No eixo x valor colocar as categorias e no eixo y vamos colocar os valores
```

```python
#Criando o banco de dados
Categorias = ['Conta A','Conta B','Conta C']
Receitas = np.array([300, 500, 700])
```

```python
# Aumentando o tamanho dos gráficos
plt.figure(figsize=(10,10))
#Plotando o gráfico de barras
plt.bar(range(0,3),Receitas)
#alterando os eixos
plt.xticks(range (0,3),Categorias)# inicio 0 até 3
plt.show()
```

## 4- SubPlots - Gráficos uns próximos aos outros

```python
#1º Criar Banco de dados
Categorias = ['Conta A','Conta B','Conta C']
Receitas = np.array([300, 450, 700])

#2º Plotar os gráficos
# Acertando o tamanho dos gráficos
plt.figure(figsize=(10,5))

# Plotando o gráfico de barras
plt.subplot(131)#posição do gráfico 1 linha 3 colunas 1 posição
plt.bar(range(0,3),Receitas)
#alterando os eixos
plt.xticks(range (0,3),Categorias)# inicio 0 até 3

#Plotando o gráfico de linhas
plt.subplot(132)#posição do gráfico 1 linha 3 colunas 2 posição
plt.plot(range(0,3),Receitas, c='red')
#alterando os eixos
plt.xticks(range (0,3),Categorias)# inicio 0 até 3


#Plotando o gráfico de Dispersão
plt.subplot(133)#posição do gráfico 1 linha 3 coluna 3 posição
plt.scatter(range(0,3),Receitas, c='green')
#alterando os eixos
plt.xticks(range (0,3),Categorias)# inicio 0 até 3
plt.show()


```

## 5- Histogramas

```python
# 1º Criando o banco de dados
a = np.random.randn(10000) *10 + 100# o random.randn -> cria valores randomicamente
a
```

```python
# 2º Plotando o histograma
plt.hist(a,7) #(array, quantidade de colunas)
plt.title("Número de casas por metros Quadrados")
plt.xlabel("Metros quadrados")
plt.ylabel("Números de Casas")
plt.show()
```
